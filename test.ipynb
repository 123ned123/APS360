```python
import torch

# 1. Check if CUDA (GPU) is available
cuda_available = torch.cuda.is_available()
print(f"Is CUDA available? {cuda_available}")

if cuda_available:
    # 2. Get the number of GPUs
    print(f"Number of GPUs: {torch.cuda.device_count()}")
    
    # 3. Get the name of the current GPU
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
    
    # 4. Get Total and Available Memory
    # Returns (free_memory, total_memory) in bytes
    free_mem, total_mem = torch.cuda.mem_get_info()
    
    print(f"Total Memory: {total_mem / 1024**3:.2f} GB")
    print(f"Free Memory: {free_mem / 1024**3:.2f} GB")
```
